## Bubble Plot
Your objective is to build a Bubble Plot that showcases the relationship between four key variables:
- Average Fare ($) Per City
- Total Number of Rides Per City
- Total Number of Drivers Per City
- City Type (Urban, Suburban, Rural)

```python

#----------------------------------------------------

# Create a new DataFrame to save the variables values
variables = pd.DataFrame()

# Average Fare ($) Per City: To take the mean, first group by city, the use mean function, finally brings the column needed
variables['Fare'] = ride_df.groupby('city').mean()['fare'] 

# Total Number of Rides Per City: First group by city, then count the different values, and point to 'ride_id'
variables['Rides'] = ride_df.groupby('city').count()['ride_id'] 

# Total Number of Drivers Per City: The variable is in the DF city
# City Type: The variable is in the DataFrame city
# After an analysis of the information, the 'ride' table has the data, and the city complements on info of the cities
# To obtain some parameters it is necesary to merge on the tables.
variables = pd.merge(variables, city_df, how = 'left', on = 'city')
variables = variables.rename(columns={'driver_count':'Driver', 'type':'City_type'})

 
```

```python
### Merge colors type, into the new DataFrame

color = pd.DataFrame({'City_type':['Urban','Suburban','Rural'],'Color_type':['orange','red','b']})
variables = pd.merge(variables, color, how = 'left', on = 'City_type')
```

## Pie Chart
In addition, you will be expected to produce the following three pie charts:
- % of Total Fares by City Type
- % of Total Rides by City Type
- % of Total Drivers by City Type

```python
# variables.drop(['Driver', 'City_type', 'Color_type'], axis = 1, inplace = True)
```

```python
variables['%Fares'] = (variables['Fare']/variables['Fare'].sum())*100
variables['%Rides'] = (variables['Rides']/variables['Rides'].sum())*100
variables['%Driver'] = (variables['Driver']/variables['Driver'].sum())*100

variables.groupby('City_type').sum()
```

```python
plt.rcParams.update({'font.size': 16})

fig1, ax1 = plt.subplots()
explode = (0.04, 0.04, 0.04)


ax1.pie(variables.groupby('City_type').sum()["%Fares"],
       labels= variables.groupby('City_type').sum().index.values,
       explode = explode, colors =['royalblue','crimson','darkorange'],
        autopct='%1.1f%%', shadow = True,
       wedgeprops = {'linewidth': 1})

ax1.set_title("% of Total Fares by City Type")
plt.show()
# ax2.pie(variables.groupby('City_type').sum()["%Fares"],
#        labels= variables.groupby('City_type').sum().index.values,
#        explode = explode, colors =['royalblue','crimson','darkorange'],
#         autopct='%1.1f%%', shadow = True,
#        wedgeprops = {'linewidth': 1})
# ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
#         shadow=True, startangle=90)
# ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# plt.show()
```

```python
fig2, ax2 = plt.subplots()
explode = (0.04, 0.04, 0.04)


ax2.pie(variables.groupby('City_type').sum()["%Rides"],
       labels= variables.groupby('City_type').sum().index.values,
       explode = explode, colors =['royalblue','crimson','darkorange'],
        autopct='%1.1f%%', shadow = True,
       wedgeprops = {'linewidth': 1},startangle=90)

ax2.set_title("% of Total Rides by City Type")
plt.show()
```
